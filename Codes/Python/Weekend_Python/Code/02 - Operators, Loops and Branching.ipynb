### Loops <a class="anchor" id="Loops"></a>
A Loop executes the sequence of statements many times until the stated condition becomes false. A loop consists of two parts, a body of a loop and a control statement. The control statement is a combination of some conditions that direct the body of the loop to execute until the specified condition becomes false. The purpose of the loop is to repeat the same code a number of times.

### For Loop  <a class="anchor" id="for-loop"></a>

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

### Example <a class="anchor" id="for-loop-Example"></a>

```python
for i in range(10):
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

### Break Statement <a class="anchor" id="Break-Statement"></a>

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "cherry":
        break
    print(x)
```

### Python supports the usual logical conditions from mathematics: <a class="anchor" id="Mathematical-Logical-Conditions"></a>

    * Equals: a == b
    * Not Equals: a != b
    * Less than: a < b
    * Less than or equal to: a <= b
    * Greater than: a > b
    * Greater than or equal to: a >= b

### Example <a class="anchor" id="Branching-Examples"></a>

```python
# Test Contidition with if
x = 2
if x>2:
    print('Greater Than 2')
else:
    print('Less Than or Equal to 2')
```

```python
# Testing Multiple Contiditions with netsted if
x = 100
if x<=10:
    print("Less than Equal 10")
elif x>10 and x<=20:
    print("10-20")
elif x>20 and x<=30:
    print("20-30")
elif x>30:
    print("Greater than 30")
else:
    print("Not a Number")
```
